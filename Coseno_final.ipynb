```python
x=[185,170,168,179,182,188]
y=[72,56,60,68,72,77]
x1=[]
y1=[]
k1=[]
k2=[]
```

<big><big><big><big><big><big><big>Funciones

d(A, B) = 1 - (A • B) / (||A|| * ||B||)

```python
def centroide1(i):
  r=(x[i]*k1[0])+(y[i]*k1[1])
  a=sqrt(pow(k1[0],2)+pow(k1[1],2))
  b=sqrt(pow(x[i],2)+pow(y[i],2))
  c=1-((r) / (a * b))
  print("c2 iter ",i," = ",c)
  return c
```

```python
def centroide2(i):
  r=(x[i]*k2[0])+(y[i]*k2[1])
  a=sqrt(pow(k2[0],2)+pow(k2[1],2))
  b=sqrt(pow(x[i],2)+pow(y[i],2))
  c=1-((r) / (a * b))
  print("c2 iter ",i," = ",c)
  return c
```

```python
def new_centroide1(i):
  k=[]
  k.append((k1[0]+x[i])/2)
  k.append((k1[1]+y[i])/2)
  k1.clear()
  return k
```

```python
def new_centroide2(i):
  k=[]
  k.append((k2[0]+x[i])/2)
  k.append((k2[1]+y[i])/2)
  k2.clear()
  return k
```
