```python
def centroide1(i):
  r=abs((x[i]-k1[0]))+abs(y[i]-k1[1])
  #print(r)
  return r
```

```python
def centroide2(i):
  r=abs((x[i]-k2[0]))+abs(y[i]-k2[1])
  #print(r)
  return r
```

Nuevo centroide

```python
def new_centroide1(i):
  k=[]
  k.append((k1[0]+x[i])/2)
  k.append((k1[1]+y[i])/2)
  k1.clear()
  return k
```

```python
def new_centroide2(i):
  k=[]
  k.append((k2[0]+x[i])/2)
  k.append((k2[1]+y[i])/2)
  k2.clear()
  return k
```
